```
#=====================================================================================================#
#        Validación Cruzada para identificar los parámetros óptimos usando Random Forest              #
#=====================================================================================================#

from pyspark.ml.regression  import RandomForestRegressor
from pyspark.mllib.evaluation import MulticlassMetrics
from pyspark.ml import Pipeline
from pyspark.ml.evaluation import RegressionEvaluator
import numpy as np

## Profundidad de los árboles
deep = range(2, 29)

## Número de árboles
arboles = [3,10,20]

## Listas vacías para llenar con los valores de RMSE y árboles
scores_1 = []
scores = []
arb = []
prof = []

## Semilla para dividir datos de prueba y entrenamiento
# semillas = np.random.randint(low = 1000, high = 5000, size = len(deep)*len(arboles) )

contador = 0
for k in deep:
  for a in arboles:
    ## Dividir datos en conjunto de prueba y entrenamiento
    # (trainSet_g1, testSet_g1) = datos2.randomSplit([0.8, 0.2], seed = semillas[contador])
  
    ## Ajuste del modelo 
    model = RandomForestRegressor(labelCol = "label", featuresCol = "features", numTrees = a, maxDepth = k, impurity = 'variance').fit(trainSet_g1)  
  
    ## Predicción en el conjunto de entrenamiento
    prediction = model.transform(trainSet_g1)
    rddPred = prediction.select('prediction', 'label').withColumn('label', prediction.label.cast(DoubleType())).rdd
  
    ## Comparar contra lo observado a través del RMSE (en datos de entrenamiento)
    evaluator = RegressionEvaluator( labelCol="label", predictionCol="prediction", metricName="rmse")
    rmse = evaluator.evaluate(prediction)
    scores.append(rmse)
  
    ## Predicción en el conjunto de prueba
    prediction_1 = model.transform(testSet_g1)
    rddPred_1 = prediction_1.select('prediction', 'label').withColumn('label', prediction.label.cast(DoubleType())).rdd
  
    ## Comparar contra lo observado a través del RMSE (en datos de prueba)
    rmse_1 = evaluator.evaluate(prediction_1)
    scores_1.append(rmse_1)
    
    ## Vector de árboles y profundidad
    arb.append(a)
    prof.append(k)
    
    ## Contador
    contador = contador + 1
  
  
error = sqlContext.createDataFrame(zip(arb, prof, scores, scores_1), schema = ['árboles','prof','error_train','error_test'])
  
display(error)
```

```
#================================================#
#      Gradient Boosted Tree Regression          #
#================================================#
from pyspark.ml.regression import GBTRegressor
from pyspark.mllib.evaluation import MulticlassMetrics
from pyspark.ml import Pipeline
from pyspark.ml.evaluation import RegressionEvaluator
import numpy as np
from pyspark.sql.functions import lit, col, sqrt

#====================# 
# Primera estimación #
#====================#
# Dividir datos en conjunto de prueba y entrenamiento
(trainSet_g1, testSet_g1) = datos2.randomSplit([0.8, 0.2], seed = 100)

#==========#
#    RF    #
#==========#
rf = RandomForestRegressor(featuresCol="indexedFeatures",impurity='variance', numTrees = 3, maxDepth = 15)

# Chain indexer and forest in a Pipeline
pipeline_rf = Pipeline(stages=[featureIndexer_t, rf])

# Train model.  This also runs the indexer.
model_rf = pipeline_rf.fit(trainSet_g1)

#==========#
#    GBTR  #
#==========#
## Ajuste del modelo
gbt = GBTRegressor(featuresCol="indexedFeatures", maxIter = 500)
  
## Chain indexer and GBT in a Pipeline ,maxDepth=j
pipeline1 = Pipeline(stages=[featureIndexer_t, gbt])

## Ajuste del modelo 
model = pipeline1.fit(trainSet_g1)
  
#==============# 
# Predicciones #
#==============# 
## Listas para llenar con los resultados
pred_monte_gbtr = model.transform(definitivo)
pred_monte_gbtr = pred_monte_gbtr.select('prediction','COD_DANE','label','IPM')

pred_monte_rf = model.transform(definitivo)
pred_monte_rf = pred_monte_rf.select('prediction','COD_DANE','label','IPM')

## Escala original de los datos
pred_monte_gbtr = pred_monte_gbtr.withColumn('suma_orig', (math.e**pred_monte_gbtr.prediction)/(1 + math.e**pred_monte_gbtr.prediction)  )
pred_monte_gbtr = pred_monte_gbtr.withColumn('suma_2_orig', pred_monte_gbtr.suma_orig**2 )
pred_monte_gbtr = pred_monte_gbtr.drop('prediction') 

pred_monte_rf = pred_monte_rf.withColumn('suma_orig', (math.e**pred_monte_rf.prediction)/(1 + math.e**pred_monte_rf.prediction)  )
pred_monte_rf = pred_monte_rf.withColumn('suma_2_orig', pred_monte_rf.suma_orig**2 )
pred_monte_rf = pred_monte_rf.drop('prediction') 

## Semilla para dividir datos de prueba y entrenamiento
iteraciones = 100
semillas = np.random.randint(low = 1000, high = 5000, size = iteraciones)
semillas = semillas.tolist()

for k in semillas:
  # Dividir datos en conjunto de prueba y entrenamiento
  (trainSet_g1, testSet_g1) = datos2.randomSplit([0.8, 0.2], seed = k )
  
  #======#
  #  RF  #
  #======#
  rf = RandomForestRegressor(featuresCol="indexedFeatures",impurity='variance', numTrees = 3, maxDepth = 15)

  # Chain indexer and forest in a Pipeline
  pipeline_rf = Pipeline(stages=[featureIndexer_t, rf])

  # Train model.  This also runs the indexer.
  model_rf = pipeline_rf.fit(trainSet_g1)
  
  #======#
  # GBTR #
  #======#  
  ## Ajuste del modelo
  gbt = GBTRegressor(featuresCol="indexedFeatures", maxIter = 500)
  
  ## Chain indexer and GBT in a Pipeline ,maxDepth=j
  pipeline1 = Pipeline(stages=[featureIndexer_t, gbt])
 

  ## Ajuste del modelo 
  model = pipeline1.fit(trainSet_g1)
  
  #==============#
  # Predicciones #
  #==============#
  #======#
  #  RF  #
  #======#
  ## Predicción en el conjunto de prueba GBTR
  pp_rf = model_rf.transform(definitivo)
  pp_rf = pp_rf.select('prediction','COD_DANE')
  pp_rf = pp_rf.withColumnRenamed("COD_DANE",'COD_DANE_pp')
  
  ## Escala original de los datos
  pp_rf = pp_rf.withColumn('pred_orig', (math.e**pp_rf.prediction)/(1 + math.e**pp_rf.prediction)  )
  pp_rf = pp_rf.withColumn('pred_2_orig', pp_rf.pred_orig**2 )
  pp_rf = pp_rf.drop('prediction')
  
  #======#
  # GBTR #
  #======#
  ## Predicción en el conjunto de prueba
  pp = model.transform(definitivo)
  pp = pp.select('prediction','COD_DANE')
  pp = pp.withColumnRenamed("COD_DANE",'COD_DANE_pp')
  
  ## Escala original de los datos
  pp = pp.withColumn('pred_orig', (math.e**pp.prediction)/(1 + math.e**pp.prediction)  )
  pp = pp.withColumn('pred_2_orig', pp.pred_orig**2 )
  pp = pp.drop('prediction') 
  
  ## Unir datos
  pred_monte_gbtr = pred_monte_gbtr.join(pp, pred_monte_gbtr.COD_DANE == pp.COD_DANE_pp, how = "left")  
  pred_monte_rf = pred_monte_rf.join(pp_rf, pred_monte_rf.COD_DANE == pp_rf.COD_DANE_pp, how = "left")  
  
  ## Sumar datos de iteraciones pasadas
  pred_monte_gbtr = pred_monte_gbtr.withColumn('suma_orig', pred_monte_gbtr.suma_orig + pred_monte_gbtr.pred_orig)
  pred_monte_gbtr = pred_monte_gbtr.withColumn('suma_2_orig', pred_monte_gbtr.suma_2_orig + pred_monte_gbtr.pred_2_orig)
  pred_monte_gbtr = pred_monte_gbtr.select('COD_DANE',"label","IPM","suma_orig","suma_2_orig")
  
  pred_monte_rf = pred_monte_rf.withColumn('suma_orig', pred_monte_rf.suma_orig + pred_monte_rf.pred_orig)
  pred_monte_rf = pred_monte_rf.withColumn('suma_2_orig', pred_monte_rf.suma_2_orig + pred_monte_rf.pred_2_orig)
  pred_monte_rf = pred_monte_rf.select('COD_DANE',"label","IPM","suma_orig","suma_2_orig")
  
## Calcular promedio GBTR
pred_monte_gbtr = pred_monte_gbtr.withColumn('media', pred_monte_gbtr.suma_orig/(iteraciones+1))
pred_monte_gbtr = pred_monte_gbtr.withColumn('var', (pred_monte_gbtr.suma_2_orig/(iteraciones+1)) - (pred_monte_gbtr.suma_orig/(iteraciones+1))**2)
pred_monte_gbtr = pred_monte_gbtr.withColumn('sd', sqrt(pred_monte_gbtr.var))
pred_monte_gbtr = pred_monte_gbtr.withColumn('cv', pred_monte_gbtr.sd/pred_monte_gbtr.media)

## Calcular promedio RF
pred_monte_rf = pred_monte_rf.withColumn('media', pred_monte_rf.suma_orig/(iteraciones+1))
pred_monte_rf = pred_monte_rf.withColumn('var', (pred_monte_rf.suma_2_orig/(iteraciones+1)) - (pred_monte_rf.suma_orig/(iteraciones+1))**2)
pred_monte_rf = pred_monte_rf.withColumn('sd', sqrt(pred_monte_rf.var))
pred_monte_rf = pred_monte_rf.withColumn('cv', pred_monte_rf.sd/pred_monte_rf.media)
```
